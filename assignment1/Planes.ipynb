```python
#bargraph for airlines

import matplotlib.pyplot as plt
airline_count = df['Airline'].value_counts()
plt.figure(figsize=(15,5))
plt.bar(airline_count.index,airline_count.values,color='g')
plt.xlabel('Airlines')
plt.ylabel('No. of Flights')
plt.title('No. of Flights for each Airline')
plt.show()
```

```python
#Calculate the Average Price for Each airlines

average_price_by_airline = df.groupby('Airline')['Price'].mean()
print(average_price_by_airline)
```

```python
#Bar Graph for average_price_by_airline

average_price_by_airline.plot(kind='bar', figsize=(15,5))
plt.xlabel('Airlines')
plt.ylabel('Average Price')
plt.title('Average Price for each Airline')
plt.show()
```
