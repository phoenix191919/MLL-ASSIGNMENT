```python
# line graph for mean of 2011-2021

import matplotlib.pyplot as plt
#calculate the mean of each year
mean_unemployment = df[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']].mean(axis=0)

plt.plot(mean_unemployment.index, mean_unemployment.values)

#set labels and title
plt.xlabel('Year')
plt.ylabel('Mean Unemployment Rate')
plt.title('Line Graph of Mean Unemployment Rate from 2010 to 2021')
plt.show()
```

```python
#all names in country column

country_names = df['country_name'].unique()

for name in country_names:
  print(name)
```

```python
#unemployment rates between continents vs mean_unemployment

import matplotlib.pyplot as plt
#groupinh the data by continent and calculate the mean unemployment rate for each year
mean_unemployment_by_continent = df.groupby('continent')[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']].mean()

for continent, mean_unemployment in mean_unemployment_by_continent.iterrows():
  plt.plot(mean_unemployment.index, mean_unemployment.values, label=continent)

#mean unemployment rate for all countries
plt.plot(mean_unemployment.index, mean_unemployment.values, label='Mean Unemployment Rate', linestyle='--')

plt.xlabel('Year')
plt.ylabel('Mean Unemployment Rate')
plt.title('Comparison of Unemployment Rates between Continents and Mean Unemployment Rate')
plt.legend()
plt.show()
```
