```python
plt.figure(figsize=(8, 6))
plt.scatter(df['income_man'], df['marriage_duration'], alpha=0.5)
plt.xlabel('Income of Man')
plt.ylabel('Duration of Marriage')
plt.title('Income of Man vs. Duration of Marriage')
plt.grid(True)
plt.show()
```

```python
#bEducation_man vs Education_women pie chart

education_man = df['education_man'].value_counts()
education_woman = df['education_woman'].value_counts()

fig, axes = plt.subplots(1, 2)
fig.set_figheight(5)
fig.set_figwidth(10)

axes[0].pie(education_man, labels=education_man.index, autopct='%1.1f%%')
axes[0].set_title('Education of Man')

axes[1].pie(education_woman, labels=education_woman.index, autopct='%1.1f%%')
axes[1].set_title('Education of Woman')

plt.show()
```

```python
#scatter plot to compare the income of the man and the income of the woman.

plt.figure(figsize=(8, 6))
plt.scatter(df['income_man'], df['income_woman'], alpha=0.5)
plt.xlabel('Income of Man')
plt.ylabel('Income of Woman')
plt.title('Income of Man vs. Income of Woman')
plt.grid(True)
plt.show()
```
